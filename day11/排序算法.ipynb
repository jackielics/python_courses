# 排序算法

冒泡排序

```python
# 冒泡排序
def bubble(seq:list)->list:
    arr = seq.copy()
    i = 0
    while i < len(arr) - 1:
        # 记录每趟最小值的位置
        min_pos = i
        j = i + 1
        while j < len(arr):
            if arr[j] < arr[min_pos]:
                min_pos = j
            j += 1
        # 开始交换
        arr[i], arr[min_pos] = arr[min_pos], arr[i]
        i += 1
    return arr

bubble([__import__('random').randint(int(-1e5), int(1e5)) for _ in range(int(1e5))])
```

插入排序

将后面的待排序序列插入到前面的已排序的序列中

```python
# 插入排序
def insert(seq:list)->list:
    arr = seq.copy()
    i = 1 # 从1到len-1
    while i < len(arr):
        j = i - 1 # i是要插入的值
        insert_value = arr[i] # 避免a[i]被覆盖
        while j >= 0:
            if insert_value < arr[j]:
                arr[j + 1] = arr[j] #
            else:
                break
            j -= 1
        arr[j + 1] = insert_value
        i += 1
    return arr
insert([__import__('random').randint(int(-1e5), int(1e5)) for _ in range(int(1e5))])
```

```python
# 选择排序
def select(seq:list)->list:
    arr = seq.copy()
    for i in range(0, len(arr) - 1): # 从0到len-2，最后一个不用移动
        im = i # 记录最小值的位置
        for j in range(i, len(arr)): 
            # 寻找最小值的位置，最后一个需要比较
            if arr[j] < arr[im]:
                im = j
        # 交换a[i]与a[im]
        arr[i], arr[im] = arr[im], arr[i]
    return arr
select([__import__('random').randint(int(-1e5), int(1e5)) for _ in range(int(1e5))])
```

快速排序

```python
# 快速排序
def quick(seq: list) -> list:
    length = len(seq)
    # 退出条件
    if length <= 1:
        return seq
    else:
        # 选取基准值
        pivot = seq.pop()
    items_greater = []
    items_lower = []
    for item in seq:
        if item > pivot:
            items_greater.append(item)
        else:
            items_lower.append(item)

    return quick(items_lower) + [pivot] + quick(items_greater)


quick([__import__('random').randint(int(-1e5), int(1e5)) for _ in range(int(1e5))])
```

堆排序

```python
def heapify(arr, n, i):
    """
    将指定节点 i 以及其子节点调整为大根堆
    :param arr: 待排序的数组
    :param n: 数组的长度
    :param i: 当前节点的索引
    """
    # 初始化最大值索引为当前节点
    largest = i
    # 左子节点索引
    l = 2 * i + 1
    # 右子节点索引
    r = 2 * i + 2
    # 如果左子节点存在且大于当前节点，更新最大值索引
    if l < n and arr[l] > arr[largest]:
        largest = l
    # 如果右子节点存在且大于当前节点，更新最大值索引
    if r < n and arr[r] > arr[largest]:
        largest = r
    # 如果最大值不是当前节点，交换节点，并递归调整其子树
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heap_sort(arr):
    """
    堆排序
    :param arr: 待排序的数组
    :return: 排序后的数组
    """
    n = len(arr)
    # 构建大根堆，(len//2-1)的位置是最后一个非叶子节点
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)
    # 逐个取出堆顶元素放到末尾，并调整堆
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]
        heapify(arr, i, 0) # 待排序的数组长度减一，直到为1
    return arr

heap_sort([__import__('random').randint(-int(1e5), int(1e5)) for _ in range(int(1e5))])
```

归并排序

```python
# 定义归并函数，输入左右两个有序序列，输出合并后的有序序列
def merge(left, right):
    result = []   # 定义一个结果序列
    i, j = 0, 0   # 定义两个游标 i 和 j，分别指向左右序列的起始位置
    while i < len(left) and j < len(right):   # 当 i 和 j 都没有超出各自序列的长度时，执行循环
        if left[i] <= right[j]:   # 如果左序列当前元素小于等于右序列当前元素
            result.append(left[i])   # 将左序列当前元素添加到结果序列中
            i += 1   # 左游标后移一位
        else:   # 如果左序列当前元素大于右序列当前元素
            result.append(right[j])   # 将右序列当前元素添加到结果序列中
            j += 1   # 右游标后移一位
    # 将剩余的元素添加到结果序列中，此时左右序列中只有一个还有剩余元素
    result += left[i:]   
    result += right[j:]
    return result   # 返回合并后的有序序列

# 定义归并排序函数，输入一个序列，输出排序后的序列
def merge_sort(seq):
    if len(seq) <= 1:   # 如果序列长度小于等于 1，则为有序序列，直接返回
        return seq
    else:
        mid = len(seq) // 2   # 将序列分为左右两部分，取中间位置
        left = merge_sort(seq[:mid])   # 递归调用归并排序函数，对左部分进行排序
        right = merge_sort(seq[mid:])   # 递归调用归并排序函数，对右部分进行排序
        return merge(left, right)   # 将排序后的左右两部分序列合并成一个有序序列，并返回结果

# 测试归并排序算法
merge_sort([__import__('random').randint(-int(1e5), int(1e5)) for _ in range(int(1e5))])
```

计数排序

用空间换时间的方式来提高排序效率，适合去重

```python
import random

def counting_sort(seq):
    """
    对输入的序列进行计数排序
    """
    max_value = max(seq)   # 获取序列中的最大值
    # 初始化计数数组，长度为最大值+1，所以不适合负数排序和浮点数排序、以及数值范围很大的排序
    count = [0] * (max_value + 1)
    for value in seq:   # 统计每个值出现的次数
        count[value] += 1 # 计数数组的索引对应序列中的值，值对应出现的次数
    result = []   # 定义结果数组
    for i in range(len(count)):   # 将计数数组转换为有序序列
        result.extend([i] * count[i])
    return result   # 返回排序后的序列

counting_sort([random.randint(0, int(1e5)) for _ in range(int(1e5))])
```

# sort与sorted的使用

基本区别

```python
# sort()的例子
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
my_list.sort(reverse=True)
print(my_list)

# sorted()的例子
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
sorted_list = sorted(my_list)
print(sorted_list)  # 输出 [1, 1, 2, 3, 3, 4, 5, 5, 5, 6, 9]
print(my_list)      # 输出 [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
```

自定义key参数

```python
names = ['J', 'Luc', 'Maggie', 'Andy']
sorted_names = sorted(names, key=lambda x: len(x))
print(sorted_names)

names.sort(key=lambda x: len(x))
print(names)
```

字典中有列表

```python
my_dict = {'a': [3, 1, 4], 'b': [5, 2, 6], 'c': [9, 7, 8]}
# .items()返回元组(key, value) 
sorted(my_dict.items(), key=lambda x: x[1][1]) # 对于每个元素，返回该元素值的第二个元素，作为排序的依据。
# my_dict.items() 
```

itemgetter与attrgetter

```python
# itemgetter 主要用于获取容器对象中指定位置的元素
from operator import itemgetter

data = [('Alice', 25), ('Bob', 20), ('Charlie', 30)]

# 按照第二个元素排序
data.sort(key=itemgetter(1))

print(data)
# 输出 [('Bob', 20), ('Alice', 25), ('Charlie', 30)]
```

```python
# attrgetter 主要用于获取容器对象中指定属性的值。
from operator import attrgetter

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f'Person(name={self.name}, age={self.age})'

data = [Person('Alice', 25), Person('Bob', 20), Person('Charlie', 30)]

# 按照年龄排序
data.sort(key=attrgetter('age'))

print(data)
# 输出 [Person(name=Bob, age=20), Person(name=Alice, age=25), Person(name=Charlie, age=30)]
```

```python
from operator import itemgetter

class MyClass:
    def __init__(self, data):
        self.data = data

    def __getitem__(self, key):
        if isinstance(key, slice):
            return [self.data[i] for i in range(*key.indices(len(self.data)))]
        elif isinstance(key, int):
            return self.data[key]
        elif isinstance(key, tuple):
            return itemgetter(*key)(self.data)
        else:
            raise TypeError("Invalid argument type")

my_obj = MyClass([1, 2, 3, 4, 5])
print(my_obj[0])  # 1
print(my_obj[1:4])  # [2, 3, 4]
print(my_obj[(0, 2, 4)])  # (1, 3, 5)
```
